```python
import cv2
from cv2 import dnn_superres
import numpy as np
import os

color_dir = 'landscapes/color'
gray_dir = 'landscapes/gray'

final_resolution = (128, 128)

# final_resolution = (400, 400)
# sr = dnn_superres.DnnSuperResImpl_create()
# model_path = "FSRCNN_x3.pb"
# sr.readModel(model_path)
# sr.setModel("fsrcnn", 3)

color_processed = []
gray_processed = []

for img_file in os.listdir(color_dir):
    print(f"now processing {img_file}")
    color_img_path = os.path.join(color_dir, img_file)
    gray_img_path = os.path.join(gray_dir, img_file)

    color_img = cv2.imread(color_img_path)
    gray_img = cv2.imread(gray_img_path)

    if color_img is None or gray_img is None:
        print(f"image {img_file} not found in either color or gray. skipping.")
        continue

    # color_img_upscaled = sr.upsample(color_img)
    # gray_img_upscaled = sr.upsample(gray_img)


    color_img_final = cv2.resize(color_img, final_resolution, interpolation=cv2.INTER_LANCZOS4)
    gray_img_final = cv2.resize(gray_img, final_resolution, interpolation=cv2.INTER_LANCZOS4)

    color_img_rgb = cv2.cvtColor(color_img_final, cv2.COLOR_BGR2RGB)

    color_processed.append(np.array(color_img_rgb))
    gray_processed.append(np.array(gray_img_final))

print(f"processing complete: {len(color_processed)} images.")
np.save("data/color_processed.npy", np.array(color_processed))
np.save("data/gray_processed.npy", np.array(gray_processed))
```
